```python
# Import the libraries.
import pandas as pd
import matplotlib.pyplot as pp
```

```python
# Load the data source into data frames.
#dataFrameConfirmedCases = pd.read_csv("time_series_19-covid-Confirmed.csv");
#dataFrameDeathCases = pd.read_csv("time_series_19-covid-Deaths.csv");
#dataFrameRecoveredCases = pd.read_csv("time_series_19-covid-Recovered.csv");
dataFrameConfirmedCases = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
dataFrameDeathCases = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
dataFrameRecoveredCases = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
dataFrameConfirmedCases
```

```python
# Melt down the various pivot tables
dataFrameConfirmedCasesMelted = pd.melt(
    dataFrameConfirmedCases,
    id_vars = [
        'Province/State',
        'Country/Region',
        'Lat',
        'Long'
    ],
    var_name = 'Date',
    value_name = 'value'
)
dataFrameConfirmedCasesMelted['classification'] = 'Confirmed Cases'
dataFrameConfirmedCasesMelted
```

```python
dataFrameDeathCasesMelted = pd.melt(
    dataFrameDeathCases,
    id_vars = [
        'Province/State',
        'Country/Region',
        'Lat',
        'Long'
    ],
    var_name = 'Date'
)
dataFrameDeathCasesMelted['classification'] = 'Deaths'
dataFrameDeathCasesMelted
```

```python
dataFrameRecoveredCasesMelted = pd.melt(
    dataFrameRecoveredCases,
    id_vars = [
        'Province/State',
        'Country/Region',
        'Lat',
        'Long'
    ],
    var_name = 'Date'
)
dataFrameRecoveredCasesMelted['classification'] = 'Recoveries'
dataFrameRecoveredCasesMelted
```

```python
# Union or join the tables row-wise. Append dataframes of like schemas together.
data_frame = dataFrameConfirmedCasesMelted.append(
    dataFrameDeathCasesMelted
)
data_frame = data_frame.append(
    dataFrameRecoveredCasesMelted
)
data_frame
```

```python
# Make sure that the Date column is being treated as a date-time data type.
data_frame['Date'] = pd.to_datetime( data_frame['Date'] )

# Make sure that dataframe is sorted by date in order since it is causing issues in pivot table.
data_frame = data_frame.sort_values('Date')
data_frame
```

```python
# Save the cleaned up, combined data set as a CSV.
data_frame.to_csv('combined_data_set.csv')
```

```python
# Form a single pivot table with all relevant data points for visualization. Sum all results.
pivot_table = pd.pivot_table(
    data_frame, 
    index = [
        'Country/Region',
        'Date'
    ],
    columns = [ 
        'classification',
    ],
    values = [
        'value'
    ],
    aggfunc = 'sum'
)
pivot_table
```

```python
# Choose which country to chart and visualize the statistics.
country = 'China'
pivot_table_country = pivot_table.xs( (country) )
#pivot_table_country
plot = pivot_table_country.plot(
    figsize  = (20, 10),
    kind = "line",
    title = "COVID-19 in " + country,
)
plot.legend(loc = "upper left")
```
